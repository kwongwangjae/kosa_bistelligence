```python
from datetime import datetime
import random
import json

# 센서 데이터 생성 함수
def temperature_sensor():
    data = {}

    data["sensor"] = "온도"
    data["value"] = round(random.uniform(20, 30), 2)
    data["datetime"] = datetime.now().strftime("%Y.%m.%d:%H:%M:%S")

    return data


# JSON 데이터 5개 생성
sensor_list = []

for _ in range(5):
    sensor_list.append(temperature_sensor())


# JSON 문자열 변환
str_json = json.dumps(
    sensor_list,
    ensure_ascii=False,
    indent=4
)


# txt 파일 저장
with open("sensor.txt", "w", encoding="utf-8") as file:
    file.write(str_json)

print("sensor.txt 저장 완료")
```
