```python
# 1~10 범위를 반복할 때 5가 되면 반복을 중지
i = 0
while True:
    i += 1
    print(i)
    if i == 5:
        break
    
```

```python
for i in range(10):
    print(i)
    if i == 5:
        break
```

```python
i = 0
total = 0
while i != 10:
    i += 1
    if i % 3 != 0:
        total += i
        continue

print(total)    
```
