```python
class Person:
    def __init__(self, name:str) -> None:
        self.name:str = name

p1 = Person("홍길동")
print(p1.name)
```

```python
class Student(Person):
    def __init__(self, name:str, sno:str) -> None:
        super().__init__(name)
        self.sno = sno

s1 = Student("홍길동", "20260103")
print(s1.name)
print(s1.sno)
```

```python
from typing import override

class A:
    def method(self):
        print("A.method() 실행")

class B:
    def method(self):
        print("B.method() 실행")

class C:
    def method(self):
        print("C.method() 실행")

class D(A, B, C):
    pass

class Parent:
    def method(self):
        print("Parent.method() 실행")

class Child(Parent):
    @override
    def method(self):
        super().method() # 100줄을 재사용
        print("Child.method() 실행 ")


c = Child()
c.method
```
