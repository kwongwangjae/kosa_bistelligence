```python
class Member:
    def __init__(self, name:str, ssn:str) -> None:
        self.__name = name
        self.__ssn = ssn

    @property
    def name(self):
        return self.__name

    @name.setter
    def name(self, name:str):
        self.__name = name

    @property
    def ssn(self):
        return self.__ssn

    @ssn.setter
    def ssn(self, ssn:str):
        if len(ssn) != 14:
            raise ValueError("주민번호 자리가 맞지 않음 ")
        self.__ssn = ssn
```

```python
member = Member("홍길동", "123456-1234567")

# 읽기 속성 사용
print(member.name)
# 쓰기 속성 사용
member.name = "홍길남"
print(member.name)


# 읽기 속성 사용
print(member.ssn)
# 쓰기 속성 사용
member.name = "654321-777777"
print(member.ssn)
```
