```python
from typing import Final

class Car:
    # 클래스 필드 (: .. 이런 방식은 권장이지 필수 강제가 아니다.)
    kind:str = "승용차"
    COMPANY: Final[str] = "현대자동차"

    # 객체 초기화
    def __init__(self) -> None:
        # 인스턴스 필드
        self.speed = 0

    def run(self) -> None:
        self.speed = 60
        print(f"{Car.COMPANY}가 {Car.speed}로 달립니다.")
```

```python
print(Car.kind)
Car.kind = "SUV"
print(Car.kind)

print(Car.COMPANY)
Car.COMPANY = "기아" #가급적 이런 방식을 쓰지 마라 
print(Car.COMPANY)
```

```python
class Car:
    # cls를 쓰려면 붙여야 한다.
    @classmethod
    def method1(cls):
        print("method1() 실행")

    @staticmethod
    def method2():
        print("method2() 실행")

    def method3():
        print("method3() 실행")
```

```python
Car.method1()
Car.method2()
Car.method3()
```

```python
from typing import Self
# 싱글톤 구현하기

# 클래스 메서드, 정적 메서드 -> 객체가 없이도 접근이 가능하다. 
# cls를 받게해야하는 경우는 클래스 메서드, cls 없어도 되면 정적 메서드로 사용
class MemberService:
    # 클래스 변수 
    _member_service = None
    _initialized = False

    # 객체 생성 역할
    def __new__(cls) -> Self:
        if cls._member_service == None:
            cls._member_service = super().__new__(cls)
        return cls._member_service

    # 객체 초기화 역할
    def __init__(self) -> None:
        if self._initialized == False:
            print("__init__() 실행해서 초기화")
            self._initialized = True

    @classmethod
    def get_instance(cls):
        return cls()
```

```python
member_service1 = MemberService()
member_service2 = MemberService()

print(member_service1 is member_service2)
```
