```python
#  __ 이런 식은 매직 메서드 라고 명칭
# __new__() : 생성자의 역할을 함 / 클래스 메서드 -> java 정적 메서드

from typing import Self
from typing import override

""" Car 클래스 정의 """
class Car:
    # 객체를 생성하는 메서드
    @override
    def __new__(cls) -> Self:
        print("__new__() 메서드 실행")
        return super().__new__(cls)

    # 객체를 초기화하는 메서드
    # 인스턴스 메서드
    @override
    def __init__(self) -> None:
        print("__init__() 메서드 실행")

```

```python
my_car = Car()
```

```python
class Car:
    @override
    def __new__(cls, name:str, year:int) -> Self:
        # 유효성 검사 필요
        if not isinstance(name, str):
            raise TypeError("name은 str 타입이어야 합니다.")
        if not isinstance(year, int):
            raise TypeError("year는 int 타입이어야 합니다.")
        return super().__new__(cls)

    @override
    def __init__(self, name:str, year:int) -> None:
        pass
```

```python
my_car = Car("그랜저", 2026)

```

```python
from typing import override

class Car:
    @override
    def __init__(self,
                 name: str,
                 year: str =  "2026",
                 price: int = 0
            ) -> None:
        # 인스턴스 필드 선언과 초기화 
        self.name = name
        self.year = year
        self.price = price
```

```python
my_car = Car("G80", "2026", 8000)
print(my_car.name)
print(my_car.year)
print(my_car.price)
```

```python
class Car:
    def __init__(self, speed:int = 0, tire:str|None = None):
        self.speed = speed
        self.tire = tire

    # 인스턴스 메서드를 정의
    def run(self) -> int:
        self.set_speed(60)
        print(f"자동차가 {self.speed}km로 달립니다.")
        return self.speed

    def set_speed(self, speed:int) -> None:
        self.speed = speed

    def change_tire(self, tire:str) -> str:
        self.tire = tire
        print(f"{self.tire}로 교체합니다.")
```

```python
my_car = Car()

current_speed = my_car.run()
my_car.set_speed(100)
print(my_car.speed)
print(my_car.change_tire("넥센"))
```
