- 문자열 인덱싱과 슬라이싱

```python
# 슬라이싱 문법: [시작 인덱스: 끝 인덱스: 스텝]
numbers = "123456789"
print(numbers[0:7:2])
print(numbers[0::2])
```

```python
# 주민번호에서 생년월일과 뒷자리 7자리를 따로 출력하시오. 
# 남자인지 여자인지 출력하시오

ssn = "123456-1234567"
print(ssn[:6])
print(ssn[7:])

print(ssn[7] == "1")
if ssn[7] == "1" or ssn[7] == "3":
    print("남자")
else:
    print("여자")
```

```python
str = "123456-38090123"

print(str.index("-"))
print(str.find("-", 4, 8))

seperater_index = str.find("-")
if seperater_index != -1:
    # 있다면
    part1 = str[:seperater_index]
    part2 = str[seperater_index:]
    print(part1)
    print(part2)
else:
    print("없다 인마!")
```
