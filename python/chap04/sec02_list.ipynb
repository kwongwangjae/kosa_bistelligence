```python
list1 : list[int] = [3, 2, 6, 7]
list2: list[int] = []

for item in list1:
    if item%2 == 0:
        list2.append(item**2)

print(list2)
```

```python
list1 : list[int] = [3, 2, 6, 7]
list2: list[int] = [ item for item in list1]

print(list2)
```

```python
# 인덱싱으로 변경 : 변경 값만 제공
# 슬라이싱으로 변경 : 리스트 형태로 변경 값을 제공
scores = [99, 95, 93, 91, 88, 84, 78, 75, 73, 70]

scores[-3:] = [0, 0, 0]
print(scores)


scores[1:2] = [100]
print(scores)
```

```python
scores = [99, 95, 93, 91, 88, 84, 78, 75, 73, 70]

scores.reverse()
print(scores)
```
