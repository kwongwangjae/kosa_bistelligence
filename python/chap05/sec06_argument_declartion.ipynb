```python
# a*값을 받을 수 있고,
def fun1(*a ,b ,c):
    pass

# *값을 받을 수 없다
def fun2(*, a, b):
    pass

# 1은 에러임
fun2(a = 1, b = 2)
fun2(b = 1, a = 2)
```

```python
# *(아스타) 뒤에는 무조건 키워드 매개변수
def fun(a, b, *, c, d = "python"):
    print(a, b, c, d)

fun(1, 2, c = 3)
```

```python
# /뒤에는 위치에 따라 값이 들어가긴 한다. 키워드 형태로도 가능(앞에는 위치 기반)
#  / 앞: 위치 고정 매개변수
#  / 뒤: 위치 또는 키워드 매개변수
def fun(a, b, /, c, d):
    print(a, b, c, d)

fun(1, 2, 3, 4)
fun(1, 2, c=3, d=4)
```

```python
# **a -> 딕셔너리

def fun(**a):
    print(type(a))
    for key in a:
        print(f"{key}: {a.get(key)}")

# a.get(key) 에러 방지 [key] 는 방지가 안됨


fun(name="파이썬", age=27)
```

```python
def fun(a, **b):
    print(a, b)

fun(1, k1 = 2, k2 = 3)
```

```python
person = {"홍길동", 30}
name, age= person
print(name, age)

person = ["홍길동", 30]
name, age = person
print(name, age)

person = {"name":"홍길동", "age":30}
```

```python
def fun(**args):
    print(args)

person = {"name":"홍길동", "age":30}

# 언패킹(key=value, ..)
fun(name = "홍길동", age = 30)
fun(**person)
```

```python
def fun(*, mid, mname, mage):
    print(mid, mname, mage)

member = {"mid": "user1", "mname": "홍길동", "mage": 30}
fun(**member)
fun(mid = "user1", mname = "홍길동", mage = 30)
```

```python
# Controller = Route
# 클래스 네임 스페이스 = 정적 멤버
# 인스턴스 네임 스페이스 = 인스턴스 멤버

# 전역 네임 스페이스에 선언된 변수
var = 1

def fun():
    # 선언 전 사용이라 에러가 남
    # print(var)
    # 함수 네임스페이스에 선언된 변수(로컬 변수)
    var = 2
    print(var)

fun()
print(var)
```

```python
var = 1

def fun():
    # 로걸 네임 스페이스로 선언 하고 사용이라 불가능
    # var = var + 1
    print(var)

fun()
print(var)
```

```python
var = 1

def fun():
    global var
    var += 1
    print(var)

fun()
print(var)
```
