```python
def score_avg(scores: list[int]) -> float:
    sum: int = 0

    while scores:
        sum += scores.pop()

    avg: float = sum / len(scores)
    return avg

scores: list[int] = [90, 80, 70]
result: float = score_avg(scores)
print(result)
```
