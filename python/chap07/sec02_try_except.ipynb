```python
while True:
    try:
        print("숫자를 입력하세요")
        x = input()
        result = 10 / int(x)
    except:
        print("0 또는 숫자가 아닌 수를 입력했습니다.")
    except EnvironmentError:
        print("에라 에라 ")
    else:
        print("결과: ,", result)
    finally:
        print("계속 여부 (y|n)")
        exit = input()
        if exit == "n":
            break
```
